# Find all granules for a given RGT

```python
import icepyx as ipx
import numpy as np
```

```python
pattern = "processed_ATL{product:2}_{datetime:%Y%m%d%H%M%S}_{rgt:4}{cycle:2}{orbitsegment:2}_{version:3}_{revision:2}.h5"
product = 'ATL09'
```

```python
# bounding box, given for Summit
coords_Summit = [-38.45918, 72.57949] # lon, lat

dN = 0.5
dE = 0.3

bbox = coords_Summit*2

for i, d in enumerate([-dE,-dN,dE,dN]):
    bbox[i] = bbox[i] + d

print(bbox)
```

```python
# determine the RGT 
rgt = 749
```

```python
region = ipx.Query(product,spatial_extent=bbox,tracks=rgt)
region.avail_granules(ids=True)
```
